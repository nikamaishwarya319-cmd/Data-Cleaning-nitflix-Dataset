```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("dirty-employee.csv")
```

```python
print(df.head())
```

```python
print("Shape:",df.shape)
```

```python
df.info()
```

```python
df=df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Unnamed: 10'],inplace=True)
```

```python
df['Age']=pd.to_numeric(df['Age'],errors="coerce")
```

```python
df['Full_Name']=df['Full_Name'].str.strip()
```

```python
df['Phone'].fillna('Not Available',inplace=True)
```

```python
df['Full_Name']=df['Full_Name'].str.strip()
```

```python
df['Gender']=df['Gender'].str.title().str.strip()
```

```python
df.dropna(subset=['Joining_Date'],inplace=True)
```

```python
df['Marks']=pd.to_numeric(df['Marks'],errors='coerce')
```

```python
df=df[(df['Age']>=18)&(df['Age']<=60)]
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Salary']=df['Salary'].astype(str).str.replace(',','',regex=False)
```

```python
df['Phone']=df['Phone'].astype(str).str.replace(r'\D','',regex=True)
```

```python
df.isnull().sum()
```

```python
df.to_csv('dirty-employee_cleaned.csv',index=True)
```
