```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("dirty-employee (1).csv")
```

```python
df.head()
```

```python
print(df.info())
```

```python
df=df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Unnamed: 10'],inplace=True)
```

```python
df['Full_Name'].fillna('Unknown',inplace=True)
```

```python
df['Age'].describe()
```

```python
df['Joining_Date']=pd.to_datetime(df['Joining_Date'],errors='coerce')
```

```python
df['Joining_Date']=df['Joining_Date'].dt.strftime("%Y-%m_%d")
```

```python
df["Salary="]=df["Salary"].replace("$","")
```

```python
df["Salary"]=df["Salary"].replace(",","")
```

```python
df["Gender"]=df["Gender"].replace({
    "M":"Male",
    "F":"Female",
    "male":"Male",
    "female":"Female"})
```

```python
df["Country"]=df["Country"].str.title()
```
