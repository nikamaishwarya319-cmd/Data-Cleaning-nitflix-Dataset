```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Attendance_Prediction (2).csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df['attendance']=pd.to_numeric(df['attendance'])
df
```

```python
df['attendance_Status']=df['attendance'].apply(lambda x:'Good'if x>=75 else 'Poor')
df
```

```python
df['attendance_Status']=pd.cut(
    df['attendance'],
    bins=[0,50,75,100],
    labels=['Low','Medium','High'],
)
df
```

```python
attendance_Status=pd.get_dummies(df['attendance'])
attendance_Status
```

```python
from sklearn.preprocessing import MinMaxScaler
Scaler=MinMaxScaler()
df['age_Scaled']= Scaler.fit_transform(df[['age']])
df







```

```python
from sklearn.preprocessing import StandardScaler
Scaler=StandardScaler()
df['age_Scaled']= Scaler.fit_transform(df[['age']])
df







```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.hist(df['attendance'], bins=10, edgecolor='black')
plt.title('attendance Distribution')
plt.xlabel('attendance (%)')
plt.ylabel('Number of Students')
plt.show()
```

```python
df['attendance'].value_counts().plot(kind='bar')

plt.title('attendance Frequency')
plt.xlabel('attendance')
plt.ylabel('Count')
plt.show()
```

```python


plt.boxplot(df['attendance'])

plt.title('attendance Box Plot')
plt.ylabel('attendance')
plt.show()
```

```python
df['attendance_Status'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('attendance Status Distribution')
plt.ylabel('')
plt.show()
```

```python
df['attendance_Status'].isnull().sum()
df['attendance_Status'].fillna(df['attendance_Status'].mode()[0],inplace=True)
df['attendance_Status'].isnull().sum()
```
